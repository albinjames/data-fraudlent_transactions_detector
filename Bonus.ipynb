👇 Evaluate a base `LogisticRegression` for Recall. Use model parameter `class_weight = 'balanced'` to ensure it deals with class imbalance.

```python
X = df.drop(columns = 'Class')
y = df['Class']
```

```python
r_scaler = RobustScaler()
r_scaler.fit(X)
X_scaled = pd.DataFrame(r_scaler.transform(X), columns = X.columns)
```

```python
log_model = LogisticRegression(class_weight='balanced', max_iter=1024)
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size=0.3,random_state=64)

cv_results = cross_validate(log_model, X_scaled, y, cv = 5, scoring=['accuracy',  #Keeping these separate
                                     'recall', 
                                     'precision',
                                     'f1'])
```

```python
pd.DataFrame(cv_results)
```

```python
base_score = pd.DataFrame(cv_results)['test_recall'].mean()
base_score
```

```python
log_model = LogisticRegression(class_weight='balanced', max_iter=1024)

log_model.fit(X_train, y_train)
y_pred = log_model.predict(X_test)
y_pred
```

```python
np.unique(y_pred, return_counts=True)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

pred_proba_df = pd.DataFrame(log_model.predict_proba(X_test))

threshold_list = [0.01, 0.04, 0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,.7,.75,.8,.85,.9,.95,.99]
for j in threshold_list:
    print ('\n******** For j = {} ******'.format(j))
    y_test_pred = pred_proba_df.applymap(lambda x: 1 if x > j else 0)
    test_recall = metrics.recall_score(y_test.to_numpy().reshape(y_test.to_numpy().size,1),
                                           y_test_pred.iloc[:,1].to_numpy().reshape(y_test_pred.iloc[:,1].to_numpy().size,1))
    print('Our testing recall is {}'.format(test_recall))

    print(confusion_matrix(y_test.to_numpy().reshape(y_test.to_numpy().size,1),
                           y_test_pred.iloc[:,1].to_numpy().reshape(y_test_pred.iloc[:,1].to_numpy().size,1)))
```

```python
# From the notes

model = LogisticRegression(class_weight='balanced', max_iter=1024)
model.fit(X_scaled,y)
def custom_predict(X, custom_threshold):
    probs = model.predict_proba(X) # Get likelihood of each sample being classified as 0 or 1
    
    expensive_probs = probs[:, 1] # Only keep expensive likelihoods (1)
    return (expensive_probs > custom_threshold).astype(int) # Boolean outcome converted to 0 or 1

y_pred = custom_predict(X=X_test, custom_threshold=0.505539) # Update predictions
print("Recall:", recall_score(y_test, y_pred)) # Rerun recall

print("Precision:", precision_score(y_test, y_pred)) # Rerun precision
print("F1 Score:", f1_score(y_test, y_pred)) # Rerun f1
```
